```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#0 - 256 valore del canale
#Intensità - Quantità di pixel col valore di quel canale

image = cv2.imread('skin/pelle8.jpg')

blu = cv2.calcHist([image], [0], None, [256], [0, 256])
rosso = cv2.calcHist([image], [1], None, [256], [0, 256])
verde = cv2.calcHist([image], [2], None, [256], [0, 256]) 

plt.figure(figsize=(10,5))
plt.subplot(2,1,1)
plt.title("Istogramma di tutti i colori")
plt.hist(blu,color="darkblue")
plt.hist(verde,color="green")
plt.hist(rosso,color="red")
plt.xlabel('Quantità di pixel')
plt.ylabel('Canale')
 
plt.subplot(2,1,2)
plt.title("Grafico di tutti i colori")
plt.plot(blu, "-b", label = "blu")
plt.plot(verde, "-g", label = "verde")
plt.plot(rosso, "-r", label = "rosso")
plt.legend(loc = 'upper left')
plt.ylabel('Quantità di pixel')
plt.xlabel('Canale')
 
plt.tight_layout()
plt.show()
```
