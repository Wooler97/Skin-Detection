```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('skin/pelle9.jpg')
color = ('b','g','r')
plt.figure()
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
    plt.ylim([0,256])
plt.show()
```
