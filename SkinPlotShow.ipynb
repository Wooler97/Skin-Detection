## Istogrammi con texture pelle

```python
skin1 = plt.imread('skin/pelle8.jpg')
graph1 = plt.imread('skin/res8.jpg')

skin2 = plt.imread('skin/pelle9.jpg')
graph2 = plt.imread('skin/res9.jpg')

skin3 = plt.imread('skin/pelle32.jpg')
graph3 = plt.imread('skin/res32.jpg')

skin4 = plt.imread('skin/pelle19.jpg')
graph4 = plt.imread('skin/res19.jpg')

skin5 = plt.imread('skin/pelle21.jpg')
graph5 = plt.imread('skin/res21.jpg')

skin6 = plt.imread('nonskin/sweat8.jpg')
graph6 = plt.imread('nonskin/res8.jpg')

skin7 = plt.imread('nonskin/sweat12.jpg')
graph7 = plt.imread('nonskin/res12.jpg')

skin8 = plt.imread('nonskin/sweat22.jpg')
graph8 = plt.imread('nonskin/res22.jpg')

skin9 = plt.imread('nonskin/sweat25.jpg')
graph9 = plt.imread('nonskin/res25.jpg')

titles = ['Texture', 'Istogramma']
images = [skin1, graph1, skin2, graph2, skin3, graph3, skin4, graph4, skin5, graph5, skin6, graph6, skin7, graph7, skin8, graph8, skin9, graph9]
plt.figure(figsize=(25,25))
for i in range(10):
    plt.subplot(5,2,i+1),plt.imshow(images[i],'gray')
    #plt.title(titles[i], fontsize = 30)
    plt.xticks([]),plt.yticks([])
plt.tight_layout()
plt.show()
```

# Confronto con i valori medi dei canali RGB tra le texture di pelle e quelle non pelle

```python
im = plt.imread('images/grafop.jpg')
im2 = plt.imread('images/grafonp.jpg')

images = [im, im2]
plt.figure(figsize=(20,10))
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray')
    plt.xticks([]),plt.yticks([])
plt.tight_layout()
plt.show()
```

La differenza tra texture in cui √® presente pelle e quelle in cui non √® presente si nota guardando i grafici rappresentanti i valori medi dei canali RGB delle foto campionate: dove √® presente la pelle nessun valore medio dei canali scende sotto alla soglia di circa 50 (nel range tra 0 e 256) e rimangono comunque molto al di sopra.
