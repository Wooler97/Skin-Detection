## Disposizione gamma di colori nello spazio RGB a seconda delle tonalit√† di pelle

```python
from matplotlib import pyplot as plt
```

```python
skin = plt.imread('skin/pelle8.jpg')
graph = plt.imread('skin/res8.jpg')

titles = ['Texture', 'Istogramma']
images = [skin, graph]
plt.figure(figsize=(25,9))
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i], fontsize = 30)
    plt.xticks([]),plt.yticks([])
plt.tight_layout()
plt.show()
```
