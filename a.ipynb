```python

import cv2
import numpy as np
from matplotlib import pyplot as plt
from tabulate import tabulate
```

```python
img = cv2.imread('skin/pelle10.jpg', cv2.IMREAD_COLOR)
blu = cv2.calcHist([img], [0], None, [256], [0, 256])
rosso = cv2.calcHist([img], [1], None, [256], [0, 256])
verde = cv2.calcHist([img], [2], None, [256], [0, 256]) 
ch1 = cv2.mean(img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,5))
plt.subplot(2,3,1)
plt.imshow(img)

plt.subplot(2,3,2)
plt.title("Istogramma di tutti i colori")
plt.hist(blu,color="darkblue")
plt.hist(verde,color="green")
plt.hist(rosso,color="red")
plt.xlabel('Quantit di pixel')
plt.ylabel('Canale')
 
plt.subplot(2,3,3)
plt.title("Grafico di tutti i colori")
plt.plot(blu, "-b", label = "blu")
plt.plot(verde, "-g", label = "verde")
plt.plot(rosso, "-r", label = "rosso")
plt.legend(loc = 'upper left')
plt.ylabel('Quantit di pixel')
plt.xlabel('Canale')
 


img2 = cv2.imread('skin/pelle9.jpg', cv2.IMREAD_COLOR)
blu = cv2.calcHist([img2], [0], None, [256], [0, 256])
rosso = cv2.calcHist([img2], [1], None, [256], [0, 256])
verde = cv2.calcHist([img2], [2], None, [256], [0, 256]) 
ch2 = cv2.mean(img2)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.subplot(2,3,4)
plt.imshow(img2)
plt.subplot(2,3,5)
plt.title("Istogramma di tutti i colori")
plt.hist(blu,color="darkblue")
plt.hist(verde,color="green")
plt.hist(rosso,color="red")
plt.xlabel('Quantit di pixel')
plt.ylabel('Canale')
 
plt.subplot(2,3,6)
plt.title("Grafico di tutti i colori")
plt.plot(blu, "-b", label = "blu")
plt.plot(verde, "-g", label = "verde")
plt.plot(rosso, "-r", label = "rosso")
plt.legend(loc = 'upper left')
plt.ylabel('Quantit di pixel')
plt.xlabel('Canale')
 
plt.tight_layout()
plt.show()
```

# Lettura dei grafici e considerazioni finali
Come possiamo notare dai plot delle texture, la differenza tra texture di pelle e quelle non-pelle consiste nella disposizione dei valori medi di ciascun canale: 
- Nelle texture di pelle la disposizione  uniforme, il rosso ha valori medi pi alti degli altri canali ed in ogni foto il verde ha sempre valori tra il rosso ed il blu.
- In quelle dove non  presente pelle invece i valori non sono uniformi, si intersecano e nelle foto prese come campione tendenzialmente si hanno valori di colore pi bassi rispetto alle texture di pelle ( < 50)
